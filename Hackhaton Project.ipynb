# Variable Explainations for df10 dataset.

1) hash_content_id: content_id watched by play<br>
2) hash_play_id: id of the play<br>
3) hash_visit_id: session_id of the play<br>
4) hash_watcher_id: watcher_id of the play<br>
5) hash_film_id: film_id of the play<br>
6) hash_event_id: id of the play (deprecated)<br>
7) is_login: boolean when user is logged_in when watch<br>
8) playback_location: where user play the video (<strong> direct | embed </strong>)<br>
9) platform: platform used to play<br>
10) play_time: time the play started<br>
11) end_time: time the play ended<br>
12) referrer: attribution of the play, source of play coming from<br>
13) average_bitrate: average bitrate of play<br>
14) bitrate_range: bitrate_range set by user to play<br>
15) total_bytes: bytes used when play<br>
16) buffer_duration: buffer_duration happen while play content<br>
17) referrer_group: group of referrer<br>
18) completed: boolean which indicate the play is complete<br>
19) utm_source: campaign source<br>
20) utm_medium: campaign source<br>
21) utm_campaign: campaign source<br>
22) player_name: engine player_name<br>
23) has_ad: indicating the play has ad or not<br>
24) flash_version:<br>
25) os_name: os_name of device use to play<br>
26) os_version: os_version of device use to play<br>
27) browser_name: browser name of device use to play<br>
28) browser_version: browser version of device use to play<br>
29) app_name: app_name used to play<br>
30) autoplay: indicating the play is autoplay when watched on certain platform<br>
31) is_premium: indicate the content played is a premium content<br>
32) app_version: version app being used<br>
33) city: city where the watcher_id play content<br>
34) play_duration: duration of play<br>
35) content_type: content type of content (<strong>livestreaming | vod | catchup</strong>)<br>
36) stream_type: when content_type is livestreaming, stream_type is used to categorize the livestream (<strong>EventStream | TVStream | RadioStream</strong>)<br>
37) title: title of content<br>
38) category_name: category_name of content<br>
39) film_title: title of film<br>
40) season_name: season_name of film<br>
41) genre_name: genre name of film (if the film is available)
